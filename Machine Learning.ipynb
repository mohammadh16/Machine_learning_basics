# 5 Stages of Data Science

- __Capture__: data extraction
- __Maintain__: data warehousing, cleansing
- __Process__: data mining, clustering/classification, data modeling, data summarization
- __Communicate__: data reporting, visualization, business intelligence, decision making
- __Analyze__: predictive analysis

What is __Machine Learning__?

the idea is, instead of writing code, you feed data to a generic algo and it builds logic based on the data given.

In __Traditional Programming__ data and program is run on the computer to produce the output
In __Machine Learning__ data and output is run on the computer to create a program


Every Machine learning algorithm has three components:
- __Representation__: decision trees, set of rules, svm, neural networks
- __Evaluation__: accuracy,prediction and recall, squared error ... .
- __Optimization__: combinatorial optimization, convex optimization, constrained optimization


__ML Algorithms by Learning Method__
   - Supervised/Inductive: we have example input and example output. like classification problem, regression problem 
   - Unsupervised: the machine learns from the data for which outcomes are not known. like clustering
   - Semi Supervised: combination of two above. it uses small amount of labeled data and a large amount of unlabeled data.

# Variance

__Sample Var vs Population Var__

Sample variance is usually greater than population variance and the N-1 in the formula is to ensure that to compensate our lack of information about the population data

# Relative Frequency vs Probability

assume that we have a bag of candies where we have 20 red candies and 80 blue candies.

lets take out a candy from the bag. the probability of the candy being red is __20/100 = 0.2__ and the probability of the candy being blue is __80/100 = 0.8__.

we iterate this action 5 times and the result is __red,red,blue,red,blue__.

so the relative frequency of the red candy is 3/5 = 0.6 and the blue candy is 2/5 = 0.4.


__Relative Frequency = Number of Successes / Total Trials__

__Probability = Happening of the desired event / total events__

also there is a difference between __Theoretical__ and __Experimental__ probability.
first is when you just calculate the probability without doing the experiment and the second one is you do the experiment and then recording the result.
